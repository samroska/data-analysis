```python
import pandas as pd
```

```python
df = pd.read_csv('./assets/mental_health.csv',header=0)
df.head()
```

```python
years = df.SurveyID.unique()
print(years)
group = {}
for year in years:
    year_df = df[df['SurveyID'] == year]
    group[str(year)] = year_df
```

```python
df_pivot = group['2014'].pivot(index='UserID', columns='questiontext', values='AnswerText').reset_index()
df_pivot
```
